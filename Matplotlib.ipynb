```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#create a line plot 
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Basic Line Plot")
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
# custumized line plot
plt.plot(x,y,color="red",linestyle="--",marker ="o")
plt.grid(True)
```

```python
#Multiple Plots
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,8,27,64,125]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1) # if there are two graphs then the position of the graphs (row,col,position)
plt.plot(x,y1,color="green")
plt.title("Plot-1")
plt.subplot(1,2,2)
plt.plot(x,y2,color="red")
plt.title("Plot-2")
```

#histograms
A histogram is a type of bar chart that shows the distribution of numerical data by grouping values into ranges (called bins or intervals).

It tells you:

How many data points fall within each range.

How the data is spread out, skewed, or concentrated.

Term	Meaning
Bins	Ranges of values (e.g., 010, 1020). Each bar represents one bin.
Frequency	How many values fall into each bin (height of the bar).
No gaps	Histogram bars touch each other because the data is continuous.

```python
data = [55, 67, 72, 85, 90, 76, 82, 60, 59, 95, 88, 73, 77, 81, 70, 66, 58, 69, 74, 91]
#createing histogram
plt.hist(data,bins=5, color = 'red',edgecolor='black')
```

```python
import pandas as pd
sales_data = pd.read_csv('Dataset.csv')
```

```python
sales_data.head()
```

```python
# PLOT TOTAL SALES BY PRODUCT
sales_data.info()
```

```python
total_sales_by_product = sales_data.groupby('Product')['Sales'].sum()
print(total_sales_by_product)
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
#PLOT SALES TREND OVER TIME
sales_trend = sales_data.groupby('Sales-Date')['Sales'].sum().reset_index()
print(sales_trend)
plt.plot(sales_trend['Sales-Date'],sales_trend['Sales'])
```
